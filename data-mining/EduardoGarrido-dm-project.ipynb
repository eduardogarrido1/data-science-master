Existen varios Barrios que todas sus entradas de Square.Meters son NA, vamos a eliminar del dataset todos los pisos que pertenecen a estos barrios.

```r
barrios_todos_NA <- df_madrid[c("Neighbourhood","Square.Meters")] %>% 
                    group_by(Neighbourhood) %>%
                    summarise(sum(na.omit(Square.Meters))) %>%
                    filter(`sum(na.omit(Square.Meters))`==0)

barrios_a_borrar <- barrios_todos_NA$Neighbourhood

df_madrid <- df_madrid %>% filter(!(Neighbourhood %in% barrios_a_borrar))
df_madrid
```

El barrio parece ser un indicador importante para los metros cuadrados de un apartamento.

Vamos a agrupar los barrios por metros cuadrados. Podemos usar una matriz de similaridad de Tukey tal y como hicimos en el curso de estadística:

```r
tky<-TukeyHSD(aov( formula=Square.Meters~Neighbourhood, data=df_madrid ))
tky.result<-data.frame(tky$Neighbourhood)
cn <-sort(unique(df_madrid$Neighbourhood))
resm <- matrix(NA, length(cn),length(cn))
rownames(resm) <- cn
colnames(resm) <- cn
resm[lower.tri(resm) ] <- round(tky.result$p.adj,4)
resm[upper.tri(resm) ] <- t(resm)[upper.tri(resm)] 
diag(resm) <- 1
library(ggplot2)
library(reshape2)
dfResm <- melt(resm)
ggplot(dfResm, aes(x=Var1, y=Var2, fill=value))+
  geom_tile(colour = "black")+
  scale_fill_gradient(low = "white",high = "steelblue")+
  ylab("Class")+xlab("Class")+theme_bw()+
  theme(axis.text.x = element_text(angle = 90, hjust = 1),legend.position="none")
```

Usando como variable de distancia: 1-resm
Dibuja un dendrograma de los diferentes barrios.


```r
md <- as.dist(1-resm)
hc <- hclust(md)
dendrograma <- as.dendrogram(hc)
hc
```

```r
par(cex=0.4)
plot(dendrograma)
abline(h=0.2, col="blue")
```

¿Que punto de corte sería el aconsejable?, ¿cuantos clusters aparecen?

```r
paste("dado el dendrograma, lo ideal sería hacer el corte en 0.2, pues, en ese punto se manifiestan 3 clusters",
      "que aportan distribuyen de manera más o menos equitativa la cantidad de barrios que hay en el dataset")
```

```r
clusters <- cutree(dendrograma,k=3)
clusters <- data.frame(as.factor(clusters))
names(clusters)="neighb_id"
clusters["Neighbourhood"]=rownames(clusters)
clusters
```

Vamos a crear una nueva columna en el dataframe df_madrid con un nuevo identificador marcado por los clusters obtenidos. Esta columna la llamaremos neighb_id

```r
df_madrid_clusters <- merge(x=df_madrid, y=clusters, by="Neighbourhood")
df_madrid_clusters
```

Vamos a crear dos grupos, uno test y otro train.

Tratamos de predecir los metros cuadrados en función del resto de columnas del dataframe.

```r
# Registros cuyo valor de Square.Meters es diferente de NA
df_madrid_sm_noNA <- df_madrid_clusters %>% filter(!is.na(Square.Meters))
paste("La cantidad de registros cuyo valor de Square.Meters es diferente de NA es:",df_madrid_sm_noNA %>% nrow())

# Parámetros inciales
set.seed(231)
muestra <- sample(1:nrow(df_madrid_sm_noNA),nrow(df_madrid_sm_noNA)*0.7)

# Creación del grupo train
df_madrid_sm_noNA.train <- df_madrid_sm_noNA[muestra, ]
paste("La cantidad de registros del grupo train es:",df_madrid_sm_noNA.train %>% nrow())

# Creación del grupo test
df_madrid_sm_noNA.test <- df_madrid_sm_noNA[-muestra, ]
paste("La cantidad de registros del grupo test es:",df_madrid_sm_noNA.test %>% nrow())
```

```r
model <- lm(Square.Meters ~ Accommodates
                            +Bathrooms
                            +Bedrooms+Beds
                            +Price
                            +Guests.Included
                            +Extra.People
                            +Review.Scores.Rating
                            +neighb_id, data = df_madrid_sm_noNA.train)
summary(model)
```

```r
# Versión mejorada del modelo
medel <- lm(Square.Meters ~ Bathrooms+Bedrooms+Price+neighb_id,
            data = df_madrid_sm_noNA.train)
summary(model)
```

```r
# Predicción del train
pred_train <- predict(model,df_madrid_sm_noNA.train, )
caret::postResample(pred_train,df_madrid_sm_noNA.train$Square.Meters)

# Predicción del test
pred_test <- predict(model,df_madrid_sm_noNA.test, )
caret::postResample(pred_test,df_madrid_sm_noNA.test$Square.Meters)
```

Mirad el histograma de los residuos sobre el conjunto de test para evaluar la calidad de vuestro modelo

```r
r_test <- df_madrid_sm_noNA.test$Square.Meters - pred_test
```

```r
plot(df_madrid_sm_noNA.test$Square.Meters,r_test)
```

```r
# Eliminamos las columnas Neighbourhood y neighb_id para quedarnos con las columnas de datos númericos
new_df_madrid <- new_df_madrid[c("Accommodates",
                                 "Bathrooms",
                                 "Bedrooms",
                                 "Beds",
                                 "Price",
                                 "Guests.Included",
                                 "Extra.People",
                                 "Review.Scores.Rating",
                                 "Latitude",
                                 "Longitude",
                                 "Square.Meters")]
new_df_madrid
```

```r
# Eliminamos los registros que poseen NA para poder hacer los cálculos
new_df_madrid_noNA <- na.omit(new_df_madrid)
new_df_madrid_noNA
```

```r
# Mostramos los autovalores
prcomp_madrid <- prcomp(new_df_madrid_noNA, center=FALSE, scale.=TRUE)
prcomp_madrid
```

```r
# Establecemos las matrices de eigenvalues y eigenvctors
matriz_eigenvalues <- prcomp_madrid$sdev
matriz_eigenvectors <- prcomp_madrid$rotation

matriz_eigenvalues
matriz_eigenvectors
```

```r
cantidad_pca <- 3
rango_pca <- prcomp_madrid$x[,1:cantidad_pca]
```

```r
apartamento <- data.frame(Accommodates=6,
                          Bathrooms=1,
                          Bedrooms=4,
                          Beds=3,
                          Price=80,
                          Review.Scores.Rating=80,
                          Guests.Included=2,
                          Extra.People=10,
                          Latitude=40.40460,
                          Longitude=-3.708392,
                          Square.Meters=43.65601)

prediccion <- matrix(predict(prcomp_madrid,apartamento)[1:cantidad_pca], nrow=1)
dist <- rep(NA,nrow(rango_pca))

for (i in 1:nrow(rango_pca)){
    dist[i] <- sum((prediccion-rango_pca[i,])^2)
}
new_df_madrid_noNA[head(order(dist),5),]
```
